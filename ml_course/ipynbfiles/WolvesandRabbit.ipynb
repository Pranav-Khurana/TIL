```
r = 10000
w = 1000

a = 0.01 
b = 0.005 
c = 0.00001 
d = 0.01

print("CASE 1 WOLVES: 1000 AND RABBITS: 10000",end='\n\n')
for day in range(1,1001):
  r = (1 + a)*r - c*r*w
  w = (1 - b)*w + c*d*r*w
  if int(day%25) == 0:
    print("Population of wolves after",day,"days is",w)
    print("Population of rabbits after",day,"days is",r,end='\n\n')
    
r = 10000
w = 500
    
print("CASE 2 WOLVES: 500 AND RABBITS: 10000",end='\n\n')
for day in range(1,1001):
  r = (1 + a)*r - c*r*w
  w = (1 - b)*w + c*d*r*w
  if int(day%25) == 0:
    print("Population of wolves after",day,"days is",w)
    print("Population of rabbits after",day,"days is",r,end='\n\n')
  
    
```
