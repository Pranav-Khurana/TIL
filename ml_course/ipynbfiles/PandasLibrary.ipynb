```
'''
Pandas Library

-> built on top of NumPy
-> contains series, data frame, panel

'''

import numpy as n
import pandas as p
```

```
'''
Series : used to create a labelled array

'''

ar = n.array([10,20,30,40,50])

ser = p.Series(ar)
print(ser)

ser = p.Series(ar,index=['q','w','e','r','t'])
print(ser)

ser = p.Series(ar)
print(ser)

ser = p.Series(data=['abc','bcd','cde'])
print(ser)

ser = p.Series(data=['abc','bcd','cde'],index=['p','q','r'])
print(ser)
```

```
'''
Slicing in series

'''

print(list(ser[-2:-1]))
```

```
'''
Data Frame 1
'''

dic = {'Id':[57,89,505],
       'Name':['Host','Friend1','Friend2'],
       'course':['ML','unkown','unknown'],
       'age':[21,20,21]
}

print(dic)

df = p.DataFrame(dic)
df
```

```
df = p.DataFrame(dic,columns=['Id','Name','course','age'])
df
```
