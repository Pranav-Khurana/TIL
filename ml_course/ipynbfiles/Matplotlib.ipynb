```
'''
Data Visualization 
-> matplotlib
-> seaborn
'''

import numpy as n
import pandas as p
import matplotlib.pyplot as l
%matplotlib inline
```

```
'''
Line Graph
-> plot(xaxis,yaxis) : used to generate a line graph

'''
l.plot([1,2,3],[4,5,1])
l.show()
```

```
'''
Adding labels and titles to a graph
'''
x = [5,8,10]
y = [12,160,600]
l.plot(x,y)
l.title('MyGraph')
l.xlabel('X AXIS')
l.ylabel('Y AXIS')
l.show()
```

```
'''
Adding styles

'''
from matplotlib import style
print(l.style.available)
```

```
'''
Using ggplot style

'''
style.use('ggplot')
x = [1,3,4,6]
y = [4,3,6,32]
a = [9,6,5,3]
b = [7,3,0,1]
l.plot(x,y,'g',linewidth=5,label='line 1')
l.plot(a,b,'c',linewidth=5,label='line 2')
l.title('newgraph')
l.xlabel('x')
l.ylabel('y')
l.grid(color='k')
l.legend()
l.show()
```

```
'''
Scatter Plot
'''

x = [1, 2, 3, 4, 5, 6, 7, 8]
y = [5 ,3, 8, 2, 9, 7, 1, 7]
l.scatter(x,y,label='scatter',color='g',s=700,marker='$PRANAV$')
l.title('Scatter Plot')
l.xlabel('X')
l.ylabel('Y')
l.show()
```

```
'''
bar graph
'''

l.bar([1,3,5,7,9,11],[4,7,3,9,2,8],label='Example One')
l.bar([2,4,6,8],[5,8,2,1],label='Example Two')
l.title('BAR GRAPH')
l.xlabel('X')
l.ylabel('Y')
l.legend()
l.show()
```

```
'''
Pie Plot

'''

slices = [7,2,2,13]
activities = ['SLEEP','EAT','PLAY','WORK']
cols = ['c','m','b','y']
l.pie(slices,labels=activities,colors=cols,startangle=90,shadow=True,explode=(0.006,0.1,0.18,0),autopct='%1.1f%%')
l.title('PIE CHART')
l.show()
```

```
'''
SubPlot

sublpot() -> used to plot two or more plots in one

figure -> it can be either horizontally or vertically
horizontal --> 2 * 1
vertical -->   1 * 2

'''

from pylab import *

l1 = [x for x in range(1,21,1)] 
l2 = [x for x in range(1,41,2)]
l3 = l1[:10:2] + [7, 14]
l4 = l2[:10:2] + [2, 4]
print(l3,l4)
```

```
subplot(1,2,1)
title("subplot(1,2,1)")
plot(l1,l2,'b--')
show()
subplot(2,1,2)
title('subplot(2,1,2)')
plot(l1,l2,'r-')
show()
```

```
subplot(2,1,1)
title("subplot(2,1,1)")
plot(l3,l4,'b--')
scatter(l3,l4)
show()
subplot(2,2,2)
title('subplot(2,2,2)')
plot(l3,l4,'r')
show()
```
